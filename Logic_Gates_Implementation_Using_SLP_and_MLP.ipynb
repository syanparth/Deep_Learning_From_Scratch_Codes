```python
import numpy as np

# Define Unit Step Function
def unitStep(v):
    return 1 if v >= 0 else 0

# Design Perceptron Model
def perceptronModel(x, w, b):
    v = np.dot(w, x) + b
    y = unitStep(v)
    return y

# Logic Functions
def AND_logicFunction(x):
    w = np.array([1, 1])
    b = -1.5
    return perceptronModel(x, w, b)

def OR_logicFunction(x):
    w = np.array([1, 1])
    b = -0.5
    return perceptronModel(x, w, b)

def NOT_logicFunction(x):
    w = -1
    b = 0.5
    return perceptronModel(x, w, b)

def NAND_logicFunction(x):
    w = np.array([-1, -1])
    b = 1.5
    return perceptronModel(x, w, b)

def NOR_logicFunction(x):
    w = np.array([-1, -1])
    b = 0.5
    return perceptronModel(x, w, b)

def XOR_logicFunction(x):
    y1 = AND_logicFunction(x)
    y2 = OR_logicFunction(x)
    y3 = NOT_logicFunction(y1)

    final_x = np.array([y2, y3])
    finalOutput = AND_logicFunction(final_x)
    return finalOutput

# Function to get user input and apply logic gate
def apply_logic_gate():
    gate = input("Enter the logic gate (AND, OR, NAND, NOR, XOR): ").strip().upper()
    x = np.array([int(input("Enter value for x1 (0 or 1): ")), int(input("Enter value for x2 (0 or 1): "))])

    if gate == "AND":
        result = AND_logicFunction(x)
    elif gate == "OR":
        result = OR_logicFunction(x)
    elif gate == "NAND":
        result = NAND_logicFunction(x)
    elif gate == "NOR":
        result = NOR_logicFunction(x)
    elif gate == "XOR":
        result = XOR_logicFunction(x)
    else:
        print("Invalid gate type!")
        return

    print(f"{gate}({x[0]}, {x[1]}) = {result}")

# Run the function
apply_logic_gate()
```
